```python
import os, sys
def getPrgFiles():
    for subdir, dirs, files in os.walk("/home/robert/Dev"):
        for f in files:
            if not f.endswith(".prg"):
                continue
            yield os.path.join(subdir,f)

for f in getPrgFiles():
    ln = open(f,'rb').read(8)[4:6]
    ln_dec = int.from_bytes(ln, byteorder='little')  # => 17
    print(f, ln_dec)
    
```
