```python
def flatten(l): return [x for y in l for x in y]

seq_src = open("../psychedelia/src/c64/sequencer_data.asm",'r').readlines()
seq_bytes = flatten([l[14:-1].split(',') for l in seq_src[1:]])
```

```python
symm = """
    ; currentSymmetrySetting: 'Current symmetry setting.'                     
    ; Possible values are 0 - 4:                                              
    ; 'NO SYMMETRY     '                                                      
    ; 'Y-AXIS SYMMETRY '                                                      
    ; 'X-Y SYMMETRY    '                                                      
    ; 'X-AXIS SYMMETRY '                                                      
    ; 'QUAD SYMMETRY   '                                                      
"""
smooth = """
    ; smoothingDelay: 'Because of the time taken to draw larger patterns speed
    ; increase/decrease is not linear. You can adjust the ‘compensating delay’
    ; which often smooths out jerky patterns. Can be used just for special FX)
    ; though. Suck it and see.'                                               
"""

print(symm)
print(f"   .BYTE {seq_bytes[0]}")

print(smooth)
print(f"   .BYTE {seq_bytes[1]}")

c = 1
for i in range(2, 255*3, 3):
    b1,b2,b3 = seq_bytes[i],seq_bytes[i+1],seq_bytes[i+2],
    position = f"""
    ; Sequencer Position {c}
    .BYTE {b1},{b2}      ; X/Y Co-ordinates: X/Y Position relative to cursor.   
    .BYTE {b3}          ; Index to pattern in pixelXPositionLoPtrArray/pixelXPositionHiPtrArray   
    """
    print(position)
    c += 1

s = i+3
for i in range(s, len(seq_bytes), 8):
    print(f"    .BYTE {','.join(seq_bytes[i:i+8])}")
    
```
