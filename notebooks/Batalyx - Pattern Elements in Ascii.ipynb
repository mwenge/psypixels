```python

patterns = [
    ("""
patternXPosArray           
        .BYTE $FF,$01,$55  
        .BYTE $FE,$02,$55  
        .BYTE $FD,$03,$55  
        .BYTE $FC,$04,$55  
        .BYTE $FB,$05,$55  
        .BYTE $FA,$06,$55  
        .BYTE $55,$55      
""".split('\n')[1:-1]
    ,"""
patternYPosArray           
        .BYTE $01,$FF,$55  
        .BYTE $FE,$02,$55  
        .BYTE $03,$FD,$55  
        .BYTE $FC,$04,$55  
        .BYTE $05,$FB,$55  
        .BYTE $FA,$06,$55  
        .BYTE $55,$55      
""".split('\n')[1:-1]
      ),
]
```

```python
def convertPatternToArray(pattern):
    pattern_array = [[int(b[-2:],16) 
                          for b in l[l.find(".BYTE"):].strip().split(',')[:-1]]
                          for l in pattern]
    pattern_array = [[b if b < 128 else b - 256
                          for b in l]
                          for l in pattern_array]
    return pattern_array

def getDimension(a):
    return (max([y for l in a for y in l]) + 
             abs(min([y for l in a for y in l if y < 0]))) + 1

def toCoords(array):
    offset = abs(min([y for l in array for y in l if y < 0]))
    return [[b + offset for b in l] for l in array]
    
ascii_patterns = []
for x_p,y_p in patterns:
    x_array, y_array =  convertPatternToArray(x_p), convertPatternToArray(y_p)
    width = getDimension(x_array)
    height = getDimension(y_array)

    x_coords = toCoords(x_array)
    y_coords = toCoords(y_array)
    rendering = [[' ' for i in range(width)] for j in range(height)]
    for i,l_x in enumerate(x_coords):
        for j,x in enumerate(l_x):
            y = y_coords[i][j]
            rendering[y][x] = str(i)
    ascii_patterns += [rendering]
```
